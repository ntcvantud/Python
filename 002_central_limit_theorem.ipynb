```python
def get_sample_means(pdf_data,data_cname,nsamples,sample_size,rseed=666):
    random.seed(rseed)
    data_array=[]
    for row in pdf_data.iterrows():
        data_array.append(row[1][data_cname])
    N=len(data_array)
    assert(sample_size<N)
    sample_means=[]
    for count in range(nsamples):
        sample=random.sample(data_array,sample_size)
        sample_means.append(np.mean(sample))
    pdf_return=pd.DataFrame(sample_means)
    pdf_return.columns=["sample_mean"]
    return pdf_return
```

```python
pdf02=get_sample_means(
    pdf_data=pdf01,
    data_cname="data",
    nsamples=1000,
    sample_size=30
)
```

```python
pdf02.head(5)
```

```python
pdf02["sample_mean"].describe()
```

```python
fig=pdf02.hist(column="sample_mean",bins=10)
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.xlim([0.0,1000.0])
plt.show()
```

```python
for i in range(1,51):
    print("Sample size = {}".format(i))
    pdf03=get_sample_means(
        pdf_data=pdf01,
        data_cname="data",
        nsamples=1000,
        sample_size=i
    )
    print("Mean of sample means = {}".format(pdf03["sample_mean"].mean()))
    fig=pdf03.hist(column="sample_mean",bins=100)
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")
    plt.xlim([0,1000])
    plt.show()
```
